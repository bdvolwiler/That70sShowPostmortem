```python
dims = (20,10)
fig, ax = plt.subplots(figsize = dims)
sns.set(font_scale = 1.15)
ax.set_title("\nThat '70s Show Episode Ratings \n", size = 30)
sns.heatmap(ax = ax, data = ratings_frame, annot=True, linewidths = 3, cmap = "Blues", cbar = True, cbar_kws = {"ticks": [6.3, 6.9, 7.6, 8.3, 8.9]})
ax.set_xlabel("\nEpisode", size = 20)
ax.set_ylabel("Season\n", size = 20)
fig.savefig( "heatmap.jpg")
```
