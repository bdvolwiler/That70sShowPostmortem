```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
seasons = pd.read_csv("number_seasons.txt")
seasons.columns = ["show", "seasons", "episodes"]
print(seasons["seasons"].mean())
that_show = pd.DataFrame(seasons.iloc[-1,:]).T
that_show
```

```python
dims = (20,5)
fig, ax = plt.subplots(figsize = dims)
sns.boxplot(data=seasons, x = "episodes", palette = "Blues")
sns.swarmplot(data=that_show, x = "episodes", color = "red", size = 15)
ax.annotate("That 70s Show", xy = (200,-.05), fontsize = 22)
plt.xlabel("")
sns.set_style("whitegrid")
plt.xticks(size = 16)
plt.title("Distribution of Episode Counts Of Top Sitcoms", size = 25)
sns.despine(left = True, bottom = True)


fig.savefig("episode distribtuion.jpg")
```

```python
dims = (20,5)
fig, ax = plt.subplots(figsize = dims)

sns.boxplot(data=seasons, x = "seasons", palette = "Blues")
sns.swarmplot(data=that_show, x = "seasons", color = "red", size = 15)
ax.annotate("That 70s Show", xy = (8.25,-.05), fontsize = 22)
plt.xlabel("")
sns.set_style("whitegrid")
plt.xticks(size = 16)
plt.title("Distribution of Season Counts Of Top Sitcoms", size = 25)
sns.despine(left = True, bottom = True)


fig.savefig("season distribtuion.jpg")
```
