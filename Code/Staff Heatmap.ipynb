```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("writer_heatmap_data.csv")
data.index = data.iloc[:,1]
data = data.iloc[:,2:]
data
```

```python
original_staff = []
for x in data.values:
    if x[0] == 1:
        original_staff.append(x *2)
    else:
        original_staff.append(x)

original_staff = pd.DataFrame(original_staff, index = data.index)
original_staff
```

```python
#original_staff = data[data["original writer"] == 1]

dims = (20,15)
fig, ax = plt.subplots(figsize = dims)
ax.set_title("\nSeason Writing Teams \n", size = 30)
#sns.heatmap(ax = ax, data = data, linewidths = 3, cmap = sns.color_palette("Blues", as_cmap=True))
sns.heatmap(ax = ax, data = original_staff.iloc[:,0:-1], linewidths = 3, cmap = sns.color_palette("Blues", as_cmap=True), cbar = False)

ax.set_xlabel("\nSeason", size = 20)
ax.set_ylabel("Writer\n", size = 20)

ax.text(6.5,-.3, "Dark Blue: Original Staff\nLight Blue: Non-Original", horizontalalignment='left', size='large', color='black', weight='semibold')
fig.savefig( "writer timeline.jpg")
```
