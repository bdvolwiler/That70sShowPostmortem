### Rating Versus Number of Seasons ( Top Trending Shows )

```python
rating22 = pd.read_csv("sns_style.csv")
```

```python
import seaborn as sns
plt.figure(figsize=(10,8))
#sns.set_theme(style="whitegrid")
sns.boxplot(x="Seasons", y="Ratings", data=rating22,palette="Blues",width=0.7,medianprops={'color':'orange'}).set_title('Average rating over number of seasons of top trending TV Shows')
sns.stripplot(x="Seasons", y="Ratings", color='grey',alpha=0.3,data=rating22);

plt.savefig('plot02.png')
```

```python
import seaborn as sns
plt.figure(figsize=(10,8))
#sns.set_theme(style="whitegrid")
sns.boxplot(x="Seasons", y="Ratings", data=rating22,color = "#407BBF", width=0.7,medianprops={'color':'black'})
#sns.stripplot(x="Seasons", y="Ratings", color='grey',alpha=0.3,data=rating22)
sns.despine(left = True, bottom = True)

plt.xticks(size = 12)
plt.yticks(size = 12)

plt.xlabel("N Seasons", size = 12)
plt.ylabel("Rating\n", size = 12)
plt.title('Average Ratings Of Shows That Run N Seasons', size = 16)

plt.savefig('boxplot seasons ran.jpg')
```
