```python
import pandas as pd

import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("writer_heatmap_data.csv")
data.index = data.iloc[:,1]
data = data.iloc[:,2:]
data
```

```python
writers = data.index
writer_churn = []

for y in range(1, data.shape[1]):
    season_gains = 0
    season_losses = 0
    for x in range(0, data.shape[0]):
        churn_number = data.iloc[x,y] - data.iloc[x,y-1]
        if churn_number == 1:
            season_gains += 1
        elif churn_number == -1:
            season_losses += 1
    writer_churn.append([season_gains, season_losses])
            
    
writer_churn = pd.DataFrame(writer_churn, columns = ["gains", "losses"])
writer_churn.index = [2,3,4,5,6,7,8]
writer_churn["net"] = writer_churn["gains"] - writer_churn["losses"]
writer_churn
```

```python
dims = (10,5)
fig, ax = plt.subplots(figsize = dims)
#ax.set_title("\nSeason Writing Teams \n", size = 30)
sns.lineplot(x = writer_churn.index, y = writer_churn["gains"], color = "green")
sns.lineplot(x = writer_churn.index, y = writer_churn["losses"], color = "red")
sns.lineplot(x = writer_churn.index, y = writer_churn["net"], color = "blue")
```
